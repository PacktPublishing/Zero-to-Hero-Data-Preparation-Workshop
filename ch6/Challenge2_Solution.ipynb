### Challenge 2: Iterative or Recursive?

```python
def sort_iterative(num_list):
    sorted_array = [num_list.pop()]
    while num_list:
        num = num_list.pop()
        i = 0
        while len(sorted_array)>i and sorted_array[i]<=num:
            i+=1
        sorted_array.insert(i,num)
    return sorted_array
```

```python
def sort_recursive(num_list):
    if not num_list:
        return num_list
    
    pivot = num_list.pop()
    left = [n for n in num_list if n<= pivot]
    right = [n for n in num_list if n> pivot]
    
    return (sort_recursive(left)+
            [pivot] +
            sort_recursive(right)
           )
```

## answers

- Sketch the orchestration of CPU, RAM, and ROM that each function:

The function `sort_iterative()` only uses CPU and RAM. The CPU has to iterate over all the numbers in values in `num_list`. In each iteration, the CPU will have to find where the value that has been popped from the num_list should be placed in `sorted_array`. This will require the CPU to have to go over the numbers in `num_list` n*n times. n here represents the number of values in `num_list`.

The function `sort_recursive()` only uses CPU and RAM as well. In the beginning, the CPU grabs the first value in `num_list` and compares that with all the other values, if they are larger than the grabbed value, they are put on the right and if they are smaller or equal to the value they are put on the left. Then the CPU is instructed to go to the left pile and does the same thing to the pile, once the left part is all sorted, the CPU will go to the right pile, and does the same thing. This way CPU recursively sorts the values, and the number of times that the CPU has to go over the `num_list` will be definitely less than n*n times.

```python
import pandas as pd
import numpy as np

n = 10000
random_df = pd.DataFrame(index=range(n), columns=['numbers'])
random_df.numbers = np.random.randint(1,10**6,n)
random_df.to_csv('random_df.csv', index=False)
```

```python
import pandas as pd

random_df = pd.read_csv('random_df.csv')
```

```python
%%time
sort_iterative(random_df.numbers.values.tolist())
```

```python
%%time
sort_recursive(random_df.numbers.values.tolist())
```
